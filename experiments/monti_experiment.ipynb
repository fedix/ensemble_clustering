```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score as sil
from sklearn.metrics import calinski_harabaz_score as ch
import numpy as np

import random

import seaborn as sns
sns.set()
%config InlineBackend.figure_format ='retina'

%cd ..
from monti import ConsensusCluster
```

```python
def monti(X, L, K, H=5, resample_proportion=0.5):
    cls = ConsensusCluster(KMeans, L, K, H, resample_proportion)
    cls.fit(X, verbose=False)
    labels = cls.predict_hierarchical()
    print(cls.bestK)
    return labels

def plot(X, labels):
    sns.scatterplot(X[:, 0], X[:, 1], hue=labels)
```

```python
X, y = datasets.make_moons(n_samples=200, noise=0.03, random_state=42)
```

```python
labels = monti(X, 2, 5, 50, 0.8)
```

```python
plot(X, labels)
```

```python
blobs, _ = datasets.make_blobs(200, centers=2, random_state=42)
```

```python
labels = monti(blobs, 2, 5, 50, 0.8)
```

```python
plot(blobs, labels)
```
